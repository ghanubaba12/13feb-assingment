```python
"""Q2. Write a python program to print Python Exception Hierarchy.
```

```python
# Define a function to print the Exception Hierarchy
def print_exception_hierarchy(exception_class, level=0):
    # Print the class name and its level
    print(' ' * level + exception_class.__name__)

    # Get the direct subclasses of the exception class
    subclasses = exception_class.__subclasses__()

    # Recursively print the hierarchy for each subclass
    for subclass in subclasses:
        print_exception_hierarchy(subclass, level + 1)

# Call the function with the base Exception class
print_exception_hierarchy(Exception)
```

```python
The LookupError class is a built-in Python exception class that serves as the base class for all lookup-related exceptions. This class defines a number of error types that can occur when we try to access an object or data structure, such as a list or dictionary, but the requested key or index does not exist.

Two common lookup-related exceptions are KeyError and IndexError.

KeyError is raised when we try to access a dictionary with a key that does not exist. For example:

css
Copy code
my_dict = {'a': 1, 'b': 2, 'c': 3}
value = my_dict['d']
This will raise a KeyError because 'd' is not a key in the dictionary.

IndexError is raised when we try to access a list or other sequence with an index that is out of range. For example:

css
Copy code
my_list = [1, 2, 3]
value = my_list[3]
This will raise an IndexError because 3 is not a valid index for a list with three elements.

By using the LookupError class as a base class for these exceptions, we can handle them in a more general way in our code. For example, we could catch any lookup-related exception with a single except block:

python
Copy code
my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['d']
except LookupError:
    print('Key not found')
This will catch both KeyError and IndexError, as well as any other lookup-related exceptions that may occur.



```

```python
"""Q5. Explain ImportError. What is ModuleNotFoundError?
```

```python
ImportError is a built-in Python exception that is raised when there is an error importing a module. This can occur for a variety of reasons, such as when the module name is misspelled, when the module does not exist or cannot be found in the Python search path, or when there is an error in the module itself.

ModuleNotFoundError is a specific subclass of ImportError that was introduced in Python 3.6. This exception is raised when a module is not found in the Python search path. It is raised when an import statement tries to import a module that does not exist or has not been installed. For example:

python
Copy code
import some_module_that_does_not_exist
This will raise a ModuleNotFoundError because there is no module with the name some_module_that_does_not_exist in the Python search path.

Before Python 3.6, an ImportError was raised for all failed imports, regardless of whether the module was not found or there was some other issue with the import process. The introduction of ModuleNotFoundError made it easier to distinguish between these two types of errors and to provide more helpful error messages for users.

In general, both ImportError and ModuleNotFoundError indicate that there is an issue with the import process and that the module could not be imported as expected.



```
